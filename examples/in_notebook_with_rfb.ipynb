```python
viewer = Viewer(show=False, ndisplay=3)
viewer.open_sample('napari', 'cells3d')
xy = np.random.randint(0, 256, (100, 2))
z = np.random.randint(0, 100, (100, 1))
viewer.add_points(np.hstack([z, xy]), opacity=0.7, size=4)
viewer
```
